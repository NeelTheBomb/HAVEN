## Analyzing Virus Metadata Resource (VMR) Database
File: VMR_MSL41.v1.20260320.csv

Downloaded on: March 25, 2026

Url: https://ictv.global/vmr

```python
import pandas as pd
import os
```

```python
dataset_csv_filepath = os.path.join(os.getcwd(), "..", "..", "..","..", "..", "input/data/vmr/VMR_MSL41.v1.20260320.csv")
```

```python
df = pd.read_csv(dataset_csv_filepath)
```

```python
df.columns
```

```python
df
```

```python
df_1 = df[~df["Virus GENBANK accession"].isna()]
df_1.shape
```

```python
df_1["Genome coverage"].value_counts()
```

```python
df_2 = df_1[df_1["Genome coverage"] != "Partial genome (non-compliant)"]
df_2.shape
```

```python
select_columns = ["Species", "Host source", "ICTV_ID", "Virus GENBANK accession"]
final_df = df_2[select_columns]
final_df
```

```python
final_df.rename(columns={
    "Host source": "source_host_category",
    "Species": "source_virus_species_name",
    "ICTV_ID": "source_virus_species_ictvid",
    "Virus GENBANK accession": "ncbi_genbank_accession" 
}, inplace=True)
```

```python
final_df["dataset_source"] = "vmr"
```

```python
final_df.to_csv(os.path.join(os.getcwd(), "..", "..", "..","..", "..", "input/data/vmr/vmr_msl41v1_processed_20260327.csv"), index=False)
```
