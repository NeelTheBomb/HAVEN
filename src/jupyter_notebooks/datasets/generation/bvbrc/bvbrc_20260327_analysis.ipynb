## Analyzing Bacterial and Viral Bioinformatics Resource Center (BV-BRC) Database
File: input/data/bvbrc/BVBRC_genome_20260327.csv

Downloaded on: March 27, 2026

Url: https://www.bv-brc.org/view/Virus/10239

```python
import pandas as pd
import os
```

```python
dataset_csv_filepath = os.path.join(os.getcwd(), "..", "..", "..","..", "..", "input/data/bvbrc/BVBRC_genome_20260327.csv")
```

```python
df = pd.read_csv(dataset_csv_filepath)
```

```python
df.shape
```

```python
df.columns
```

```python
df
```

```python
df_1 = df[~df["GenBank Accessions"].isna()]
df_1.shape
```

```python
df_1["Host Common Name"]
```

```python
df_1["Genome Status"].value_counts()
```

```python
df_2 = df_1[df_1["Database"] == "GenBank"]
df_2.shape
```

```python
df_2["DetectionMethod"].value_counts()
```

```python
df_2
```

```python
df_2.columns
```

```python
select_columns = ["Host", "Virus", "HostTaxID", "VirusTaxID", "NCBIAccession"]
final_df = df_2[select_columns]
final_df
```

```python
final_df.rename(columns={
    "Host": "source_host_species_name",
    "Virus": "source_virus_species_name",
    "HostTaxID": "source_host_species_taxid",
    "VirusTaxID": "source_virus_species_taxid",
    "NCBIAccession": "ncbi_genbank_accession" 
}, inplace=True)
```

```python
final_df["dataset_source"] = "virion"
```

```python
final_df.to_csv(os.path.join(os.getcwd(), "..", "..", "..","..", "..", "input/data/virion/20260326/virion_processed_20260327.csv"), index=False)
```
